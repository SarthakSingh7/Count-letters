## Part 1: Python | Remove all characters except letters and numbers

link: https://www.geeksforgeeks.org/python-remove-all-characters-except-letters-and-numbers/


```python
import re

# initialising string
ini_string = input("Enter the story: ")

# printing initial string
print ("initial string : ", ini_string)

result = re.sub('[\W_]+', '', ini_string)

# printing final string
print ("final string", result)
```

```python
# converting all the letters to lowercase

lower_case = result.lower()

print(lower_case)
```

## Part  2 How many time each letter is repeated in a big story

```python
import string

d = {}  # creating a dictionary

#print(type(d))
s= string.ascii_lowercase
num = [str(i) for i in range(0,10)]
print(num)
for i in s:
    d[i] = 0
for i in num:
    d[i] = 0
    
print(d)

for j in lower_case:
    d[j] = d[j] + 1
print(d)    

    
```

```python
import pandas as pd

df = pd.DataFrame(list(d.items()),columns = ['Letters','Value']) 

print(df)
```
